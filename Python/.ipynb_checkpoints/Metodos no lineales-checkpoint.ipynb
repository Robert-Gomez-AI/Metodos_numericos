```python
import numpy as np
from IPython.display import clear_output
import time 
```

```python
import matplotlib.pyplot as plt
```

# Metodos lineales
## Ecuacion a solucionar
$$x^{2}-2=0$$

```python
def funcion(x):
    return x**2 -2
```

```python
def derivada(x):
    return 2*x
```

## Metodo de la bisecci贸n

```python
def biseccion(funcion,a0,b0,iteraciones,tol_error):
    
    a,b=a0,b0
    error=funcion(a)
    X=[]
    for i in range(iteraciones):
        c=(a+b)/2
        time.sleep(1)
        clear_output(wait=True)
        print("Intervalo inicial [{},{}],Tolerancia={} ,Iteraciones={}".format(a0,b0,tol_error,iteraciones))
        print("Iteraci贸n : {} ,Intervalo=({},{}), x={},error:{}".format(i+1,a,b,c,error))
        fc=funcion(c)
        if fc==0 or np.abs(fc)<tol_error:
            break
        if funcion(a)*fc<0:
            b,error=c,fc
        elif funcion(b)*fc<0:
            a,error=c,fc
        X.append([a,b])
    return X
        
```

```python
X0=biseccion(funcion,1,2,20,2e-7)
```

```python

plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(10,5))
i=0
for a,b in X0:
    ax.scatter((a,b),(funcion(a),funcion(b)),c='r',alpha=1/np.sqrt((len(X0)-i)))
    i+=1
X_axis=np.linspace(1,2,200)
Y_axis=funcion(X_axis)
ax.plot(X_axis,Y_axis,c='g',alpha=0.3)
ax.axhline(linewidth=1, color='0')
# ax.axvline(linewidth=1, color='0')
ax.grid(color='w', linestyle='solid')

ax.set_title('Metodo de la bisecci贸n')
```

## Metodo de regula-farsi

```python
def regula_farsi(funcion,a0,b0,iteraciones,tol_error):
    a,b=a0,b0
    error=funcion(a)
    X=[]
    for i in range(iteraciones):
        
        c=(a*funcion(b)-b*funcion(a))/(funcion(b)-funcion(a))            
        time.sleep(0.3)
        clear_output(wait=True)
        print("Intervalo inicial [{},{}],Tolerancia={} ,Iteraciones={}".format(a0,b0,tol_error,iteraciones))
        print("Iteraci贸n : {} ,Intervalo=({},{}), x={},error:{}".format(i+1,a,b,c,error))
        fc=funcion(c)
        if fc==0 or np.abs(fc)<tol_error:
            break
        if funcion(a)*fc<0:
            b,error=c,fc
        elif funcion(b)*fc<0:
            a,error=c,fc
        X.append([a,b,c])
    return X
        
```

```python
X1=regula_farsi(funcion,1,2,20,2e-7)
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(15,10))
i=0
for a,b,c in X1:
    opacity=1/np.sqrt(len(X1)-i)
    ax.scatter((a,b,c),(funcion(a),funcion(b),funcion(c)),c='green',alpha=opacity)
    ax.plot((a,b),(funcion(a),funcion(b)),c='b',alpha=opacity)
    i+=1
X_axis=np.linspace(1.3,2,200)
Y_axis=funcion(X_axis)
ax.plot(X_axis,Y_axis,c='g',alpha=1)
ax.axhline(linewidth=1, color='0')
# ax.axvline(linewidth=1, color='0')
ax.grid(color='w', linestyle='solid')

ax.set_title('Metodo de regula-farsi')
```

## Metodo de Newton

```python
def Newton(funcion,derivada,x0,iteraciones,tol_error):
    x=x0
    error=np.abs(funcion(x0))
    values=[]
    for i in range(iteraciones):
        values.append(x)
        time.sleep(0.3)
        clear_output(wait=True)
        print("Punto inicial {},Tolerancia={},Iteraciones={}".format(x0,tol_error,iteraciones))
        print("Iteraci贸n : {} , x={},error:{}".format(i+1,x,error))
        if error<tol_error:
            print("Completado")
            break
        x=x-funcion(x)/derivada(x)
        error=np.abs(funcion(x))
    return values
```

```python
X3=Newton(funcion,derivada,2,20,1e-5)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
i=0
for a in X3:
    opacity=1/np.sqrt(len(X3)-i)
    ax.scatter(a,funcion(a),c='green',alpha=opacity)
    i+=1
X_axis=np.linspace(1.3,2,200)
Y_axis=funcion(X_axis)
ax.plot(X_axis,Y_axis,c='g',alpha=1)
ax.axhline(linewidth=1, color='0')
# ax.axvline(linewidth=1, color='0')
ax.grid(color='w', linestyle='solid')
ax.set_title('Metodo de Newton')
```
